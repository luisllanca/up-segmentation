```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pandas as pd
import os
```

### Que datos tenemos??

```python
snakes = [x for x in os.listdir("../data/") if "perimFactor_1" in x and "z049" in x]
```

```python
img = cv2.imread("../data/T2_t0002_ch00_z049.tif")
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
fig,ax = plt.subplots()
kernel = np.ones((5,5),np.uint8)
#img_gray = cv2.dilate(img_gray,kernel)
ax.imshow(img_gray,cmap="Reds")
ax.axis("off")

for namedir in snakes:
    snake = pd.read_csv(os.path.join("..","data",namedir),sep="        ",header=None,engine="python")
    x,y=snake[0],snake[1]
    ax.plot(x,y)
    fig.canvas.draw()

ax.invert_yaxis()
ax.set_title(f"Size: {img_gray.shape}")
```
